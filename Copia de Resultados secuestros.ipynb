```python
import matplotlib.pyplot as plt
import numpy as np
# Ajustamos el código para hacer los círculos 10 veces más grandes según la solicitud.

# Colores para las categorías
colors = ['orange', 'red', 'green', 'black']

# Iniciar la figura de Matplotlib
fig, ax = plt.subplots(figsize=(20, 3))

# Punto de inicio de la barra
start = 0
# Tamaño del icono representativo (podemos usar un círculo como icono simplificado)
# Aumentamos el tamaño del icono por un factor de 10
icon_size = total_personas

# Dibujar cada categoría en la misma línea
for i, (categoria, cantidad) in enumerate(categorias.items()):
    end = start + cantidad
    ax.scatter(np.arange(start, end), [0]*cantidad, s=icon_size * 2.4, color=colors[i], label=f'{categoria} ({cantidad})')
    start = end

# Ajustes de visualización
ax.set_xlim(0, total_personas)
ax.set_ylim(-1, 1)
ax.axis('off')  # Sin ejes para este gráfico

# Leyenda
ax.legend(scatterpoints=1, loc='upper center', ncol=len(categorias), fontsize=12, bbox_to_anchor=(0.5, 1.25))

# Título
plt.title('Distribución de Resultados de Secuestros', fontsize=16, pad=40)

plt.show()
```
